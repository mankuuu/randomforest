<h3><kbd>n_estimators</kbd></h3>
<p>Controls the numbers of decision trees in random forests</p>
<p>Increasing the estimators increases the randomness in data. <b>Randomness helps in reducing  overfitting. More randomness Less overfitting</b></p>

```python
def n_estimator_fun(i):
    model = RandomForestClassifier(n_jobs=-1, random_state = 42, n_estimators= i ).fit(X_train,train_targets)
    return {
             'number' : i ,
             'train_acc' : model.score(X_train , train_targets) ,
             'val_acc' : model.score(X_val , val_targets) ,
    }
```

```python
estimator_df = pd.DataFrame(n_estimator_fun(i) for i in range(10,200,10))
```

```python
estimator_df.sort_values('val_acc' , ascending=False)
```

```python
plt.plot(estimator_df['number'] , estimator_df['val_acc'])
```
